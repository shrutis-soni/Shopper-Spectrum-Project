# **Project Name**    - Customer Segmentation and Product Recommendations in E-Commerce


# **Project Summary -**

The rapid growth of the e-commerce industry has resulted in the generation of massive amounts of customer transaction data every day. This data contains valuable information about customer purchasing behavior, preferences, spending patterns, and product demand. Businesses can leverage this information to improve customer satisfaction, increase sales, and make data-driven decisions. The primary objective of this project, titled "Shopper Spectrum: Customer Segmentation and Product Recommendations in E-Commerce," is to analyze customer transaction data, identify distinct customer groups, and develop an intelligent product recommendation system.

The project begins with collecting and understanding an online retail transaction dataset containing information such as Invoice Number, Product Code, Product Description, Quantity Purchased, Invoice Date, Unit Price, Customer ID, and Country. Data preprocessing is performed to ensure data quality by removing missing customer records, excluding cancelled transactions, and eliminating invalid entries with negative or zero quantities and prices. This step ensures that the analysis is based on accurate and reliable data.

A major component of the project is Customer Segmentation using RFM (Recency, Frequency, Monetary) Analysis. Recency measures how recently a customer made a purchase, Frequency represents how often the customer purchases, and Monetary indicates the total amount spent by the customer. These metrics are calculated for each customer and standardized using data scaling techniques. Clustering algorithms such as K-Means are then applied to group customers with similar purchasing behaviors. The optimal number of clusters is determined using techniques like the Elbow Method and Silhouette Score. Based on the RFM characteristics, customers are categorized into segments such as High-Value Customers, Regular Customers, Occasional Customers, and At-Risk Customers. These segments enable businesses to design targeted marketing campaigns and customer retention strategies.

The second major component is the Product Recommendation System. An item-based collaborative filtering approach is used to recommend products similar to a selected product. A customer-product interaction matrix is created, and cosine similarity is calculated between products. When a user enters a product name, the system identifies and recommends the top five most similar products based on customer purchase patterns. This enhances the shopping experience by providing personalized recommendations and increasing cross-selling opportunities.

To make the solution user-friendly and accessible, a Streamlit web application is developed. The application consists of two modules: a Product Recommendation Module that suggests similar products and a Customer Segmentation Module that predicts the customer segment based on Recency, Frequency, and Monetary values. The project demonstrates the practical application of machine learning, clustering, recommendation systems, feature engineering, and data visualization in solving real-world e-commerce challenges and supporting business growth through data-driven decision-making.

## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

Null Hypothesis (H0): Average revenue is equal across countries. Alternative Hypothesis (H1): Average revenue differs significantly across countries. Hypothesis testing helps determine whether observed differences are statistically significant.

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

H1: Quantity directly impacts Revenue
More products purchased → higher revenue

H2: Low-priced items are purchased in higher quantity
Customers prefer cheaper products in bulk

H3: Revenue is dominated by few countries
A small number of countries contribute most sales


#### 2. Perform an appropriate statistical test.

```python
# Perform Statistical Test to obtain P-Value
from scipy.stats import pearsonr

df['Revenue'] = df['Quantity'] * df['UnitPrice']
df_clean = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]

corr, p_value = pearsonr(df_clean['Quantity'], df_clean['Revenue'])

print("Correlation:", corr)
print("P-value:", p_value)
```

##### Which statistical test have you done to obtain P-Value?

Test: ANOVA (One-Way F Test)

##### Why did you choose the specific statistical test?

p < 0.05 → Significant relationship
p ≥ 0.05 → Not significant

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

#### 2. Perform an appropriate statistical test.

```python
# Perform Statistical Test to obtain P-Value
from scipy.stats import pearsonr

corr, p_value = pearsonr(df_clean['UnitPrice'], df_clean['Quantity'])

print("Correlation:", corr)
print("P-value:", p_value)
```

##### Which statistical test have you done to obtain P-Value?

##### Why did you choose the specific statistical test?

I applied Pearson correlation to test relationships between numerical variables like Quantity, Unit Price, and Revenue. For categorical comparison like revenue across countries, I used ANOVA to check if there are statistically significant differences between groups. The results help validate business hypotheses with statistical evidence rather than just visual analysis.

### Hypothetical Statement - 3

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

H3: Revenue is dominated by few countries
A small number of countries contribute most sales

#### 2. Perform an appropriate statistical test.

```python
# Perform Statistical Test to obtain P-Value
from scipy.stats import f_oneway

# Select top 3 countries for testing (avoid too many groups)
top_countries = df_clean['Country'].value_counts().head(3).index

group1 = df_clean[df_clean['Country'] == top_countries[0]]['Revenue']
group2 = df_clean[df_clean['Country'] == top_countries[1]]['Revenue']
group3 = df_clean[df_clean['Country'] == top_countries[2]]['Revenue']

f_stat, p_value = f_oneway(group1, group2, group3)

print("F-statistic:", f_stat)
print("P-value:", p_value)
```

##### Which statistical test have you done to obtain P-Value?

p < 0.05 → Revenue differs significantly across countries
p ≥ 0.05 → No significant difference

##### Why did you choose the specific statistical test?

I applied Pearson correlation to test relationships between numerical variables like Quantity, Unit Price, and Revenue. For categorical comparison like revenue across countries, I used ANOVA to check if there are statistically significant differences between groups. The results help validate business hypotheses with statistical evidence rather than just visual analysis.

```python
# Handling Outliers & Outlier treatments
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.boxplot(df['Quantity'])
plt.title("Outliers Check - Quantity")
plt.show()
```

```python
Q1 = df['Quantity'].quantile(0.25)
Q3 = df['Quantity'].quantile(0.75)

IQR = Q3 - Q1

outliers = df[
    (df['Quantity'] < Q1 - 1.5 * IQR) |
    (df['Quantity'] > Q3 + 1.5 * IQR)
]

print("Number of outliers:", len(outliers))
```

```python
#capping technique
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

df['Quantity'] = df['Quantity'].clip(lower, upper)
```

```python
# transform of log
df['Quantity_log'] = np.log1p(df['Quantity'])
```

```python
# replacing with median
median = df['Quantity'].median()

df.loc[
    (df['Quantity'] < lower) | (df['Quantity'] > upper),
    'Quantity'
] = median
```
